#Esimation of Pi value using Monte Carlo Method

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
n = 10000
x = np.random.uniform(-1, 1, size = (n, 2))
inside = x[np.sqrt(x[:,0]**2 + x[:,1]**2) <= 1]
estimate = 4* len(inside)/len(x)
print('estimate : {}'.format(estimate))
plt.figure(figsize=(8,8))
plt.scatter(x[:,0], x[:,1], s = 1, c = 'red')
plt.scatter(inside[:,0], inside[:,1], s = 1, c = 'green')
plt.show()
```
